```python
import numpy as np
import pandas as pd

import warnings
warnings.filterwarnings('ignore')
```

```python
customer = pd.DataFrame({'customer_id' : np.arange(6),
                         'name' : ['철수', '영희', '길동', '영수', '수민', '동건'],
                         '나이' : [40, 20, 21, 30, 31, 18]})
```

```python
customer
```

```python
orders = pd.DataFrame({'customer_id' : [1,1,2,2,2,3,3,1,4,9],
                       'item' : ['치약', '칫솔', '이어폰', '헤드셋', '수건', '생수', '수건', '치약', '생수', '케이스'],
                       'quantity' : [1,2,1,1,3,2,2,3,2,1]})
```

```python
orders
```

```python
pd.merge(customer, orders, on='customer_id')
```

```python
pd.merge(customer, orders, on='customer_id', how='inner')
```

```python
pd.merge(customer, orders, on='customer_id', how='left')
```

```python
pd.merge(customer, orders, on='customer_id', how='right')
```

```python
pd.merge(customer, orders, on='customer_id', how='outer')
```

```python
cust1 = customer.set_index('customer_id')
order1 = orders.set_index('customer_id')
```

```python
cust1.head()
```

```python
order1.head()
```

```python
pd.merge(cust1, order1, left_index=True, right_index=True)
```

```python
pd.merge(cust1, orders, left_index=True, on='customer_id')
```

```python
merge_df = pd.merge(customer, orders, on='customer_id')
```

```python
merge_df.sort_values(by='quantity', ascending=False)
```

```python
merge_df[(merge_df['name'] == '영희')].sort_values(by='quantity', ascending=False)
```

```python
pd.merge(customer, orders, on='customer_id').groupby('item').sum()
```

```python
dd = pd.merge(customer, orders, on='customer_id').groupby(['name', 'item']).sum().sort_values(by='quantity', ascending=False)
```
